# SVD

Дальше мы работаем с SVD. Для этого устанавливаем surprise 

Подгружаем массивы данных в тетрадку

Обучаем меодель SVD методом cross validation разбивая быорку на 5 отрезков 


```python
pip install surprise
```

```python
from surprise import SVD
from surprise import Dataset, Reader
from surprise.model_selection import cross_validate
```

```python
reader=Reader(rating_scale=(1,5))
data = Dataset.load_from_df(movies, reader)
```

```python
model = SVD()
```

```python
cross_validate(model, data, measures=['MSE', 'MAE'], cv=8, verbose=True)
```
